#**Implementação da Transformada Discreta de Fourier - DFT (Método Matricial)**

### Importação de Bibliotecas:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Função para Calcular a DFT:
São necessários dados do sinal de entrada (x) e da quantidade de amostras (N). As posições do sinal (tamanho L do sinal de entrada ) também devem ser consideradas.
Como visto em sala, devem atender as seguintes condições:

1.  x[n] tem L valores,   0 <= n <= L
2.  N >= L
3.  A resolução da DFT (N) deve ser uma potência de 2.



```python
def calcular_DFT(x, N):
    """
    Parâmetros:
    x (array): Sinal de entrada.
    N (int): Resolução da DFT.
    """
    # Condições de tamanho do sinal de entrada e da DFT
    L = len(x)
    l = int(np.log2(N))
    if L > N:
        print("O tamanho do sinal de entrada (L) não pode ser maior que o tamanho da DFT (N).")
        return None
    if N != 2 ** l:
        print("A resolução da DFT (N) deve ser uma potência de 2.")
        return None

    # Criação da matriz de coeficientes complexos W
    W = np.zeros((N, L), dtype=complex)
    for n in range(N):
        for k in range(L):
            W[n, k] = np.exp(-2 * np.pi * n * k * 1j / N)

    # Cálculo da DFT
    x_t = x.reshape(-1, 1)  # Transpor o sinal para formato de vetor coluna
    DFT = np.dot(W, x_t)


    return DFT # Coeficientes da DFT
```

### Dados de entrada:

```python
# Solicitação dos dados de entrada ao usuário
x = np.array([1, 1, 0, 0])  # Exemplo de sinal de entrada
N = int(input("Digite a resolução da DFT (N deve ser uma potência de 2): "))  # Resolução da DFT desejada
```

### Cálculo da DFT usando o método matricial:

```python
# Cálculo da DFT
DFT = calcular_DFT(x, N)

# Plotagem dos resultados
if DFT is not None:
    # Plotagem da amplitude da DFT
    plt.figure(figsize=(10, 5))
    plt.subplot(2, 1, 1)
    plt.stem(np.abs(DFT))
    plt.title("Amplitude da DFT")
    plt.xlabel("k")
    plt.ylabel("|X[k]|")
    plt.grid(True)

    # Plotagem da fase da DFT
    plt.subplot(2, 1, 2)
    plt.stem(np.angle(DFT))
    plt.title("Fase da DFT")
    plt.xlabel("k")
    plt.ylabel("Fase(X[k])")
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
